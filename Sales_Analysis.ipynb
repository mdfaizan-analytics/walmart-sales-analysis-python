```python
Store_Sales = df.groupby("Store")["Weekly_Sales"].sum()
```

```python
pd.options.display.float_format = '{:,.2f}'.format
```

```python
# Create a bar chart of the Store Sales
Store_Sales.plot(kind='bar', figsize=(15, 6), color='royalblue', edgecolor='black')

# Add professional titles and labels
plt.title('Total Weekly sales by store', fontsize=16, fontweight='bold')
plt.xlabel('Store Number', fontsize=12)
plt.ylabel('Total Sales ($)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the chart cleanly
plt.tight_layout()
plt.show()
```

```python
# Sort Stores by Sales in descending order and show the top 5
Top_Stores = Store_Sales.sort_values(ascending=False).head(5)
print("--- TOP 5 STORES BY TOTAL SALES ---")
print(Top_Stores)
```

```python
# Save the chart as a PNG picture in your folder
plt.figure(figsize=(15,6))
Store_Sales.plot(kind='bar', color='royalblue', edgecolor='black')
plt.title('Total Weekly Sales by Store', fontsize=16, fontweight='bold')
plt.xlabel('Store Number', fontsize=12)
plt.ylabel('Total Sales ($)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# This is the line that saves it!
plt.savefig('Store_Sales_chart.png', dpi=300) 
plt.show()
```

```python
# Sort the Store Sales data in descending order
Sorted_Sales = Store_Sales.sort_values(ascending=False)

print("--- TOP 5 HIGHEST PERFORMING STORES ---")
print(Sorted_Sales.head(5))

print("\n--- BOTTOM 5 LOWEST PERFORMING STORES ---")
print(Sorted_Sales.tail(5))
```
